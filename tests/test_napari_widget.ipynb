```python
import matplotlib.pyplot as plt
import numpy as np
import flimtools
import tifffile
```

```python
#img=flimtools.readBin('TileScan_003_s1_z1_seq1.bin')
img=tifffile.imread('TileScan_003_s1_z1_seq1_crop_8bit.tif')
```

```python
gimg,simg,sumimg=flimtools.calcGSimg(img,45,1)
```

```python
gshist,gbins,sbins,gthresh,sthresh,xthreshcoords,ythreshcoords=flimtools.getGSHist(gimg,simg,sumimg,smsigma=3,thresh=3)
```

```python
#need this magic to get interactivity
%matplotlib notebook
fig=flimtools.showHist(gshist,gbins,sbins,points=np.array([[0.58,0.39],[0.21,0.15]]),ptsizes=[0.1,0.1])
```
